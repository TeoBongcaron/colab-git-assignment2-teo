```python
# Add Python Function

# Multiplication function
def multiply_numbers(a, b):
    return a * b
print(multiply_numbers(3, 4))
```

```python
# Add Python Function

# Division Function
def divide_numbers(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"
print(divide_numbers(10, 2))
```
